# Cosmic Star Formation Rate
In this notebook, we plot the three main cooling functions in the interstellar medium (ISM) at temperatures below $10^5$K. The information to create this plots was taken from Figure 9.6 from Chapter 9 of *Interstellar and Intergalactic Medium* by Ryden & Pogge, 2021, Cambridge University Press. This notebook is a modified version of the one presented in the [repository of the book](https://github.com/CambridgeUniversityPress/Interstellar-and-Intergalactic-Medium/blob/main/Chapter9/Fig9_6_CosmicSFR.ipynb). Additional we have used the data from the observational results of [Madau & Dickinson 2014, ARA&A, 52, 415](https://ui.adsabs.harvard.edu/abs/2014ARA%26A..52..415M),  Equation 15 and Table 1. We include the data from Rowan-[Robinson et al. 2016](https://ui.adsabs.harvard.edu/abs/2016MNRAS.461.1100R/abstract), [Novak et al. 2017](https://ui.adsabs.harvard.edu/abs/2017A%26A...602A...5N/abstract) and [Gruppioni et al. 2020](https://ui.adsabs.harvard.edu/abs/2020A%26A...643A...8G/abstract). First, we import the Python packages.

```python
import warnings

from astropy.cosmology import Planck13, z_at_value
from astropy.table import Table
import astropy.units as u

import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator

import numpy as np

warnings.filterwarnings('ignore', category=UserWarning, append=True)
warnings.filterwarnings('ignore', category=RuntimeWarning, append=True)
```

## Standard Plot Format
Setup the standard plotting format and make the plot.

```python
# graphic aspect ratio = width/height
ASPECT = 16.0/9.0

# Text width in inches - don't change, this is defined by the print layout
TEXTWIDTH = 5.14 # inches

# output format and resolution
FIG_FMT = 'pdf'
DPI = 300

# Graphic dimensions
PLT_W = DPI*TEXTWIDTH
PLT_H = PLT_W/ASPECT
AX_FONTSIZE = 10
LAB_FONTSIZE = 8
LWIDTH = 0.5
AX_PAD = 3
WINCHES = TEXTWIDTH
HINCHES = WINCHES/ASPECT

# LaTeX is used throughout for markup of symbols, Times-Roman serif font
plt.rc('text', usetex=True)
plt.rc('font', **{'family':'serif','serif':['Times-Roman'],'weight':'bold','size':'16'})

# Font and line weight defaults for axes
matplotlib.rc('axes', linewidth=LWIDTH)
matplotlib.rcParams.update({'font.size':AX_FONTSIZE})

# axis and label padding
plt.rcParams['xtick.major.pad'] = f'{AX_PAD}'
plt.rcParams['ytick.major.pad'] = f'{AX_PAD}'
plt.rcParams['axes.labelpad'] = f'{AX_PAD}'
```

## Cosmic Star Formation Rate
Compute the cosmic SFR using the parametric fit of Madau & Dickinson 2014, equation 15:
\begin{equation}
   \psi(z) = 0.015\frac{(1+z)^{2.7}}{1+\left[(1+z)/2.9\right]^{5.6}} {\rm M_{\odot} yr^{-1} Mpc^{-3}}
\end{equation}

```python
z = np.linspace(0.0, 9.0, 501)
sfr = 0.015*((1+z)**2.7)/(1+((1+z)/2.9)**5.6)
```

Then we read the SFR density data from other works from the Table 1

```python
MDdata = Table.read('../Data/MadauDickinson2014_data.txt', format='ascii')
UVgroup = MDdata['Type'] == 'UV'
IRgroup = MDdata['Type'] == 'IR'
AddData = Table.read('../Data/AdditionalSFRD.txt', format='ascii')
RR2016 = AddData['Reference'] == 'RowanRobinson2016'
N2017 = AddData['Reference'] == 'Novak2017'
G2020 = AddData['Reference'] == 'Gruppioni2020'
```

##### Figure
We corrected the SFRD values from G2020 and N2017 because they have a different IMF. The value of 1.7 is assumed from the Novak+17 paper.

```python
%matplotlib inline

fig, ax = plt.subplots()

fig.set_dpi(DPI)
fig.set_size_inches(WINCHES, HINCHES, forward=True)

ax2 = ax.twiny()
age_ax = [1, 4, 6, 8, 9, 10, 11, 12, 12.5, 13]*u.Gyr
ageticks = [z_at_value(Planck13.lookback_time, age) for age in age_ax]

ax.errorbar(MDdata['z'][UVgroup], MDdata['SFRD'][UVgroup],
            xerr=MDdata['Deltaz'][UVgroup],
            yerr=(MDdata['low_SFRD'][UVgroup], MDdata['high_SFRD'][UVgroup]),
            fmt='o', label='UV data MD2014')
ax.errorbar(MDdata['z'][IRgroup], MDdata['SFRD'][IRgroup],
            xerr=MDdata['Deltaz'][IRgroup],
            yerr=(MDdata['low_SFRD'][IRgroup], MDdata['high_SFRD'][IRgroup]),
            fmt='o', label='IR data MD2014')
ax.errorbar(AddData['z'][RR2016], AddData['SFRD'][RR2016],
            xerr=(AddData['low_z'][RR2016], AddData['high_z'][RR2016]),
            yerr=(AddData['low_SFRD'][RR2016], AddData['high_SFRD'][RR2016]),
            fmt='^', label='Rowan-Robinson et al. 2016', mfc='white')
ax.errorbar(AddData['z'][N2017], np.log10(10**AddData['SFRD'][N2017]*1.7),  # IMF chage
            xerr=(AddData['low_z'][N2017], AddData['high_z'][N2017]),
            yerr=(AddData['low_SFRD'][N2017], AddData['high_SFRD'][N2017]),
            fmt='s', label='Novak et al. 2017', mfc='white')
ax.errorbar(AddData['z'][G2020], np.log10(10**AddData['SFRD'][G2020]*1.7),  # IMF chage
            xerr=(AddData['low_z'][G2020], AddData['high_z'][G2020]),
            yerr=(AddData['low_SFRD'][G2020], AddData['high_SFRD'][G2020]),
            fmt='>', label='Gruppioni et al. 2020', mfc='white')
ax.plot(z, np.log10(sfr), '-', color='black', label=r'Madau \& Dickinson 2014 fit',
        lw=1.0, zorder=10)

ax.set_ylim(-2.4, -0.4)
ax.set_xlim(-0.1, 10.2)
ax.set_xticks(np.arange(0, 11))

# From https://github.com/jchowk/pyND/plotting/plotting.py
ax2.set_xticks(ageticks)
ax2.set_xticklabels(['{:g}'.format(age) for age in age_ax.value])
ax2.set_xlim(ax.get_xlim())
ax2.xaxis.set_minor_locator(NullLocator())
ax2.set_xlabel('Lookback time [Gyr]')

ax.set_xlabel(r'z (redshift)', fontsize=AX_FONTSIZE)
ax.set_ylabel(
    r'SFR density [M$_{\odot}$ yr$^{-1}$ Mpc$^{-3}$]', fontsize=AX_FONTSIZE)
ax.legend(prop={'size': 6}, loc='upper right')
plt.savefig('../Figures/SFRdensity.pdf',
            bbox_inches='tight', facecolor='white')
```
