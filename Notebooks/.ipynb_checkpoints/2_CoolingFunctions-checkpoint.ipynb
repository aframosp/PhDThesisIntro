# Cooling functions
In this notebook, we plot the three main cooling functions in the interstellar medium (ISM) at temperatures below $10^5$K. The information to create this plots was taken from Figure 1.10 from Chapter 1 of *Interstellar and Intergalactic Medium* by Ryden & Pogge, 2021, Cambridge University Press. This notebook is a modified version of the one presented in [repository of the book](https://github.com/CambridgeUniversityPress/Interstellar-and-Intergalactic-Medium/blob/main/Chapter1/Fig1_FGH.ipynb).
First, we import the Python packages.

```python
import warnings
import math

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore',category=UserWarning, append=True)
```

## Standard Plot Format
Setup the standard plotting format and make the plot.

```python
# graphic aspect ratio = width/height
ASPECT = 16.0/9.0

# Text width in inches - don't change, this is defined by the print layout
TEXTWIDTH = 5.14 # inches

# output format and resolution
FIG_FMT = 'pdf'
DPI = 300

# Graphic dimensions
PLT_W = DPI*TEXTWIDTH
PLT_H = PLT_W/ASPECT
AX_FONTSIZE = 10
LAB_FONTSIZE = 8
LWIDTH = 0.5
AX_PAD = 3
WINCHES = TEXTWIDTH
HINCHES = WINCHES/ASPECT

# LaTeX is used throughout for markup of symbols, Times-Roman serif font
plt.rc('text', usetex=True)
plt.rc('font', **{'family':'serif','serif':['Times-Roman'],'weight':'bold','size':'16'})

# Font and line weight defaults for axes
matplotlib.rc('axes', linewidth=LWIDTH)
matplotlib.rcParams.update({'font.size':AX_FONTSIZE})

# axis and label padding
plt.rcParams['xtick.major.pad'] = f'{AX_PAD}'
plt.rcParams['ytick.major.pad'] = f'{AX_PAD}'
plt.rcParams['axes.labelpad'] = f'{AX_PAD}'
```

## FGH model
Definitions for the Field, Goldsmith, and Habing (FGH) multi-phase interstellar medium model [Field, Goldsmith, & Habing 1969, ApJ, 155, L149](https://ui.adsabs.harvard.edu/abs/1969ApJ...155L.149F).

```python
MINT = 10.0
MAXT = 20000.
minLogT = math.log10(MINT)
maxLogT = math.log10(MAXT)
logT = np.linspace(minLogT, maxLogT, num=1001)

def estimate_cooling(x_e, ltemp):
    """Definition to iterate in fractional ionisation"""
    tem = 10.0**ltemp
    xfac = x_e/0.001
    t_h = 1.18e5  # hydrogen excitation temperature in K
    t_c = 91.2     # carbon excitation temperature in K
    t_o = 228.0    # oxygen excitation temperature in K

    # Lyman-alpha cooling
    cool_lya = 6e5*(xfac/np.sqrt(tem/1e4))*np.exp(-t_h/tem)

    # Carbon cooling
    coolc_e = 3.1*(xfac/np.sqrt(tem/100.0))*np.exp(-t_c/tem)
    coolc_h = 5.2*((tem/100.0)**0.13)*np.exp(-t_c/tem)
    coolc = coolc_e + coolc_h
    # Oxygen cooling
    coolo = 4.1*((tem/100.0)**0.42)*np.exp(-t_o/tem)
    # Total cooling
    cool_tot = (cool_lya + coolc + coolo)
    return (cool_tot, cool_lya, coolo, coolc, coolc_e, coolc_h)
```

##### Figure

```python
%matplotlib inline

fig, axs = plt.subplots(1, 2, sharey=True, sharex=True)

fig.set_dpi(DPI)
fig.set_size_inches(WINCHES, HINCHES, forward=True)

XLABEL = r'Temperature [K]'

for ax in axs:
    ax.tick_params('both', length=6, width=LWIDTH, which='major',
                   direction='in', top='on', right='on')
    ax.tick_params('both', length=3, width=LWIDTH, which='minor',
                   direction='in', top='on', right='on')
    ax.set_xscale('log')
    ax.set_xticklabels(['10', '100', '1000', '10$^{4}$'])
    ax.set_xticks([10, 100, 1000, 1.0e4])
    ax.set_yscale('log')
    ax.set_yticks([1.0E-30, 1.0E-29, 1.0E-28,
                  1.0E-27, 1.0e-26, 1.0e-25, 1.0e-24])
    ax.set_yticklabels(['$10^{-30}$', '10$^{-29}$', '10$^{-28}$',
                       '10$^{-27}$', '10$^{-26}$', '10$^{-25}$', '10$^{-24}$'])
    ax.set_xlabel(XLABEL)

# Limits

MINCOOL = 1.0e-29  # erg cm^3 s^-1
MAXCOOL = 1.0e-24

# Labels

YLABEL = r'$\Lambda$ [erg cm$^3$ s$^{-1}$]'

plt.xlim(MINT, MAXT)

plt.ylim(MINCOOL, MAXCOOL)

axs[0].set_ylabel(YLABEL)

# Plot the total and individual cooling functions

for ifi, frac_ion in enumerate([0.01, 0.001]):
    coolings = estimate_cooling(frac_ion, logT)
    axs[ifi].set_title('x=%.3f' % frac_ion, fontdict={'fontsize': 10})
    axs[ifi].plot(10**logT, 1.0e-27*coolings[0], '-', color='k',
                  lw=2, zorder=10, label=r'$\Lambda_{\rm{Total}}$')
    axs[ifi].plot(10**logT, 1.0e-27*coolings[3], '-', color='C0',
                  lw=1, zorder=10, label=r'$\Lambda_\textup{C\,\textsc{\lowercase{II}}}$')
    axs[ifi].plot(10**logT, 1.0e-27*coolings[1], '-', color='C2',
                  lw=1, zorder=10, label=r'$\Lambda_{\mathrm{Ly}\alpha}$')
    axs[ifi].plot(10**logT, 1.0e-27*coolings[4], '--', color='C0',
                  lw=1, zorder=10, label=r'$\Lambda_\textup{C\,\textsc{\lowercase{II}}}^{e}$')
    axs[ifi].plot(10**logT, 1.0e-27*coolings[2], '-', color='C1',
                  lw=1, zorder=10, label=r'$\Lambda_\textup{O\,\textsc{\lowercase{I}}}$')
    axs[ifi].plot(10**logT, 1.0e-27*coolings[5], ':', color='C0',
                  lw=1, zorder=10, label=r'$\Lambda_\textup{C\,\textsc{\lowercase{II}}}^{H}$')

plt.subplots_adjust(wspace=0, hspace=0)

axs[1].legend(ncol=3, prop={'size': 6}, handletextpad=0.2,
              columnspacing=1, loc='upper center')
plt.savefig('../Figures/Cooling_functions.pdf', bbox_inches='tight', facecolor='white')
```
